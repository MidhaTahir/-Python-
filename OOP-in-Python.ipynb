# Static and Instance Variables

```python
#Static and Instance Variables
class Dinosaur:
    eyes = 2 #static variable or Class variable
    def __init__(self,teeth):
        self.teeth = teeth #teeth is an instance variable
        
ryna = Dinosaur(5)
dyna = Dinosaur(7)

#just used different formatting options to print for fun
print("ryna eyes: {}".format(ryna.eyes))
print("dyna eyes: {}".format(dyna.eyes))
print(f"ryna teeth: {ryna.teeth} ")
print(f"dyna teeth: {dyna.teeth} ")

ryna.teeth = 6 #teeth of ryna is changed but not for dyna (as it is instance variable)
print(f"ryna teeth: {ryna.teeth} ")
print(f"dyna teeth: {dyna.teeth} ")

Dinosaur.eyes = 4 #accessed it from class and both obj eyes got changed (static variable)
print(f"ryna eyes changed: {ryna.eyes} ")
print("dyna eyes also changed : {}".format(dyna.eyes))

ryna.eyes = 3 
print(f"ryna eyes changed: {ryna.eyes} ")
print("dyna eyes not changed : {}".format(dyna.eyes))
```

```python
class Test:
    instvar = 0
    staticvar = 0
    def __init__(self):
        self.instvar += 1
        Test.staticvar += 1 #static variable is called from Class 
        
t1 = Test()
print("t1 Static Variable:{}".format(t1.staticvar))
print("t1 Instance Variable:{}".format(t1.instvar))
t2 = Test()
print("t2 Static Variable:{}".format(t2.staticvar))
print("t2 Instance Variable:{}".format(t2.instvar))
print("t1 Static Variable printing again :{}".format(t1.staticvar))
#static variable is shared same for all obj
```

# INHERITANCE

```python
# INHERITANCE
class Employee:
    def __init__(self,empno,empname,salary,designation):
        self.empno = empno
        self.empname = empname
        self.salary = salary
        self.desig = designation
    def showEmp(self):
        print(f'The employee name {self.empname}, with employee no {self.empno} and salary {self.salary} is {self.desig}')
              
class Staff(Employee):
    def __init__(self,empno,empname,salary,designation,comm):
        self.commission = comm
        super().__init__(empno,empname,salary,designation)

emp = Staff('12','Hashy','2000000000','CEO','400000')
emp.showEmp()
print(f'Commission: {emp.commission}')
```

# METHOD OVERRIDING 

```python
# METHOD OVERRIDING 
class Employee:
    def __init__(self,empno,empname,salary,designation):
        self.empno = empno
        self.empname = empname
        self.salary = salary
        self.desig = designation
    def showEmp(self):
        print(f'The employee name {self.empname}, with employee no {self.empno} and salary {self.salary} is {self.desig}')
              
class Staff(Employee):
    def __init__(self,empno,empname,salary,designation,comm):
        self.commission = comm
        super().__init__(empno,empname,salary,designation)
    def showEmp(self): #method override
        super().showEmp()
        print(f'Commission is {emp.commission}')

emp = Staff('12','Hashy','2000000000','CEO','400000')
emp.showEmp()
```

# Multiple Inheritance

```python
# Multiple Inheritance
class Father:
    def fatherProperity(self):
        print("Father's Property")
class Mother:
    def motherProperity(self):
        print("Mother's Property")
class Child(Father,Mother):
    def childProperty(self):
        print("Child's Property is derived from: ")
        super().fatherProperity()
        super().motherProperity()

ch = Child()
ch.childProperty()
```

# Multilevel Inheritance

```python
# Multilevel Inheritance
class GrandP:
    def propLand(self):
        print("The land property from GP is used for farming")
class Parents(GrandP):
    def propHome(self):
        print("Property Home constructed by Parents")
class Child(Parents):
    def propVeh(self):
        print("Child has property vehicle")

c1 = Child()
c1.propLand()
c1.propHome()
c1.propVeh()
```

# Multilevel MEthod Overriding

```python
# Multilevel MEthod Overriding
class GrandP:
    def propLand(self):
        print("The land property from GP is used for farming")
class Parents(GrandP):
    def propHome(self):
        print("Property Home constructed by Parents")
class Child(Parents):
    def propVeh(self):
        print("Child has property vehicle")
    
    def propLand(self): #method overriden 
        print("The land property from GP is used for commercial")

c1 = Child()
c1.propLand()
c1.propHome()
c1.propVeh()
```

# MRO

```python
#Method Resolution Order (MRO)
class X: pass
class Y: pass
class Z: pass

class A(X,Y): pass
class B(Y,Z): pass

class N(A,B,Z): pass

for obj in N.mro():
    print(obj)
    
# it will first search first in N then left order A, with A in X. then with N it searches B then B left (Y) since object is not on left og both X Y it will access Z from N then object from Z as object is on left to Z 
```

#  ABSTRACT CLASSES

1) We can't define the object of abstract classes
2) In order to use abstract class we need to inherit another class from abstract class
3) The derived class which inherits the abstract class must provide the definition of all the abstract methods present in abstract class
4) when to use abstract class ... when we know the method is reqd but we can't define them same like.

```python
from abc import ABC,abstractmethod

class Institute(ABC): #inheriting ABC class
    def __init__(self):
        print(type(self).__name__,"details: ") #class name 
        #print(type(self),"details: ") #<class '__main__.Institute'> details: 
    def courseOffered(self):
        print("Cources offered are : C,C++,Java,Python")
        
    @abstractmethod
    def address(self): pass
    
    
# inst = Institute() ---> Type Error
#TypeError: Can't instantiate abstract class Institute with abstract methods address
class MidsMIT(Institute):
    def courseOffered(self):
        print("Courses Offered: Python, DataScience, Database")
        
    def address(self):
        print("Address: Karachi")
        
class OnlineMIT(Institute):
    def address(self):
        print("Address MidsMIT@lms.com")

m = MidsMIT()
m.courseOffered()
m.address()

h = OnlineMIT()
h.courseOffered()
h.address()
```

# POLYMORPHISM

```python
#Polymorphism
class English:
    def greet(self,name):
        print("Good Morning " + name)
    
class French:
    def greet(self,name):
        print("Bonjour " + name)
    
def greetings(language,name): #method .. common interface
    language.greet(name)
    
eng = English()
french = French()

greetings(eng,"Mids")
greetings(french,"Hash")
```

# Operating Overloading

```python
class Point:
    def __init__(self,x,y):
        self.X = x
        self.Y = y
        
    def __add__(self,other):
        x = self.X + other.X
        y = self.Y+ other.Y
        return Point(x,y)
    
p1 = Point(10,20)
p2 = Point(5,10)

result = p1 + p2
print(result)
#print(p1+p2) #TypeError: unsupported operand type(s) for +: 'Point' and 'Point'
```

```python
# When we have requirment to view obj as string we use __str__
class Point:
    def __init__(self,x,y):
        self.X = x
        self.Y = y
        
    def __add__(self,other):
        x = self.X + other.X
        y = self.Y+ other.Y
        return Point(x,y)
    
    def __str__(self):
        return f'{self.X,self.Y}'
    
p1 = Point(10,20)
p2 = Point(5,10)

result = p1 + p2
print(result)
```

```python
# When we have requirment to view obj as string we use __str__
class Point:
    def __init__(self,x,y):
        self.X = x
        self.Y = y
        
    def __add__(self,other):
        x = self.X + other.X
        y = self.Y+ other.Y
        return Point(x,y)
    
    def __str__(self):
        return f'{self.X,self.Y}'
    
    def __lt__(self,other):
        p1 = self.X + self.Y
        p2 = other.X + other.Y
        return p1 < p2
    
p1 = Point(10,20)
p2 = Point(15,10)

result = p1 + p2
print(result)

flag = p1 < p2
print("Is point p1 is lesser than p2? ", flag)
```
